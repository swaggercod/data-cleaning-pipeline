```python
import pandas as pd
import numpy as np

print("‚úÖ Libraries loaded successfully!")
```

```python
# Load raw data
df = pd.read_csv('../data/raw/ecommerce_raw_data.csv')

print(f"üìä Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
df.head(10)
```

```python
print("üìã Data Types:")
print(df.dtypes)
print("\n" + "="*50)
print(f"Total columns: {len(df.columns)}")
```

```python
print("‚ùå Missing Values Count:")
missing = df.isnull().sum()
print(missing[missing > 0])  # Sadece null olanlarƒ± g√∂ster

print(f"\nüìä Missing Percentage:")
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
print(missing_pct[missing_pct > 0])
```

```python
print("üîç Unique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
print("üì¶ Category Values:")
print(df['category'].value_counts(dropna=False))
# dropna=False ‚Üí null deƒüerleri de say
```

```python
print("üí∞ Price Statistics:")
print(df['price'].describe())

print("\n‚ö†Ô∏è Negative Prices:")
print(df[df['price'] < 0]['price'].count())

print("\n‚ö†Ô∏è Outliers (Price > 1000):")
print(df[df['price'] > 1000]['price'].count())
```

```python
print("üîÑ Duplicate Rows:")
duplicates = df.duplicated().sum()
print(f"Total duplicates: {duplicates}")

print("\nSample duplicate:")
print(df[df.duplicated(keep=False)].head(4))
# keep=False ‚Üí hem orijinali hem kopyasƒ±nƒ± g√∂ster
```
